#### **Exercise 1:**
Give students in the `example.csv` file a bonus of 5 for all the courses. Find a way to replace all the values that are now greater than 30 by 30. Save the new data to a file. Make a plot with histograms for every course in the example file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read file
df = pd.read_csv('./data/example.csv', index_col=0)

# Apply bonus for all the courses
df = df + 5

# Replace values greater than 30 by 30
df[df > 30] = 30

# Save the new data to a file
df.to_csv('./data/example_bonus_applied.csv')
```

```python
# Plot histogram for each course
df.hist()
plt.show()
```

#### **Exercise 2:**

 - Read in the `cars.csv` file as a dataframe
 - Remove the row with the index 3, which contains a quote in the Model column.
 - Add a new column called Price with the following values: [8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800]
 - Find the average Length of all the vehicles.
 - Find the median Price of all the vehicles.
 - Create a new dataframe that only contains the rows where the Price is greater than 7000.
 - Sort the dataframe in descending order by the Price column.


```python
# Read file
df = pd.read_csv('./data/cars.csv')
print(len(df))

# Add price column
df['Price'] = [8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800]

# Remove the third row
df.drop(2, axis='rows', inplace=True)

# Print average of `Length`
print('Mean of length:', df['Length'].mean())

# Print median of `Price`
print('Median of price:', df['Price'].median())

# Create new DataFrame
df_new = df[ df['Price'] > 7_000 ]

# Sort by Price column descendingly
print('Printing Price column descendingly...')
print(df_new.sort_values(by='Price', ascending=False))
```

#### **Exercise 3:**
 - create a random matrix with numpy
 - save it to file
 - read it back with numpy
 - calculate some statistcs on it with numpy 
 - read the same file with pandas
 - calculate the same statistics on it with pandas

```python
# Creating a random matrix
mat = np.random.random((3,4))

# Save it to file
filename = './data/random_matrix.txt'
np.savetxt(filename, mat)

# Read it again
mat = np.genfromtxt(filename)

# Calculate statistics with NumPy
print('Mean, std, max, min of each column:')
for i in range (len(mat[0])):
    print(mat[:, i].mean(), mat[:, i].std(), mat[:, i].max(), mat[:, i].min())

# Read with Pandas
df = pd.read_csv(filename, sep='\\s+', header=None)

# Print stats with Pandas
print('Stats with Pandas:')
print(df.describe())
```
